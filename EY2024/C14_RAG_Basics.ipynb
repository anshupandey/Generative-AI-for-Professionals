```python
import os
os.environ["OPENAI_API_VERSION"] = api_version
os.environ["AZURE_OPENAI_ENDPOINT"] = azure_endpoint
os.environ["AZURE_OPENAI_API_KEY"] = api_key
os.environ["AZURE_OPENAI_API_VERSION"] = api_version
```

```python
from langchain_openai import AzureOpenAIEmbeddings

embeddings = AzureOpenAIEmbeddings(azure_deployment=embedding_model_name)
```

```python
v1,v2,v3 = embeddings.embed_documents(["I like python programming, its a good programming language.",
                                    "Python is an amazing programming language, i like it a lot.",
                                    "Hello World, how are you?"])
```

# Vector DB

```python
import chromadb
chroma_client = chromadb.Client()
```

```python
collection = chroma_client.create_collection(name="my_collection2")
```

```python
collection.add(
    embeddings=[[1.2, 2.3, 4.5], [6.7, 8.2, 9.2]],
    documents=["Python is a programming language", "langchain is an framework"],
    metadatas=[{"source": "my_source"}, {"source": "my_source"}],
    ids=["id1", "id2"]
)
```

```python
results = collection.query(
    query_embeddings=[[1.1, 1.1, 2.1]],
    n_results=2
)
results
```
